```python
import pandas as pd
```

```python
current_data = pd.read_csv("cleaned_data_extended.csv", sep=',')
current_data.head()
```

```python
population = pd.read_csv("PopulationEurope.csv", sep=';')
population = pd.melt(population, id_vars=['Country'])
population['value'] = population['value'].astype('str').str.replace(r".", r"", regex=False)
population = population.rename(columns={'variable': 'Year', 'value': 'Population'})
population['Population'] = population['Population'].astype(str).astype(int)
population['Year'] = population['Year'].astype(str).astype(int)
population.head()
```

```python
merged_data_population =  current_data.merge(population,on=["Country", "Year"])
```

```python
merged_data_population["Plastic_waste_generated_per_person"] = (merged_data_population["Plastic_waste_generated"] / merged_data_population["Population"]) *1000
```
